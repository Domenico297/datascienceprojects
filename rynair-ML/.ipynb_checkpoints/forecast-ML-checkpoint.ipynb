```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('SXF-CIA_v2.dat',header=None)  #read prices

date=pd.date_range('2019-09-06', periods=len(df), freq='D') # creates data

daytodep=np.arange(len(df))/30  # creates days (or months if /30) to departure
daytodep=daytodep.astype(int)

dayofweek=[]   # creates the days of the week (1 weekend, 0 nonweekend)
kk=5
for i in range(len(df)):
    if kk>=5:
        dayofweek.append(1)  #weekend
    if kk<5:
        dayofweek.append(0)  #noweekend
    if kk<7:  
        kk=kk+1
    elif kk>=7:
        kk=1
        

# adding all the informations to the dataframe
df.index=date
df['daytodep']=daytodep
df['dayofweek']=dayofweek
df['holidays']=0

# changing the name of dataframe columns
df.columns=['Price','Daytodeparture','Dayofweek','Holidays']

# adding public holidays
df.loc['2019-10-03','Dayofweek']=1
#df.loc['2019-10-04':'2019-10-18','Dayofweek']=1
df.loc['2019-10-31','Dayofweek']=1
df.loc['2019-12-25','Dayofweek']=1
df.loc['2019-12-26','Dayofweek']=1
df.loc['2020-01-01','Dayofweek']=1
df.loc['2020-04-10','Dayofweek']=1
df.loc['2020-04-13','Dayofweek']=1

df.dropna(inplace=True)

df
```

```python
# check of holidays  (must be equal to 5 - 1 dropping out Christmas the 25th of December, no flithgs on that day)

df['Holidays'].sum()
```

```python
# preparing the dataset for the machine learning

X=np.array(df.iloc[:,1:3])
y=np.array(df.iloc[:,0])

plt.hist(y,bins=[0,25,50,75,100])
plt.show()
```

```python
plt.plot(y)
plt.xlabel('Days to departure')
plt.ylabel('Price €')
plt.show()
```

```python
# pivoting the dataframe

dfpivot=df.pivot_table(columns='Dayofweek',index='Daytodeparture',values='Price')

dfpivot.plot()
plt.legend(['During the week','Weekend'])
plt.show()
```

```python
dfpivot['Daytodeparture']=dfpivot.index
dfpivot.reset_index

# reshaping to have the data for Machine Learning
dfavg=pd.melt(dfpivot,id_vars='Daytodeparture')

dfavg
```

```python
# new training set, after reshaping

X2=np.empty([dfavg.shape[0],dfavg.shape[1]-1])

X2[:,0]=dfavg.iloc[:,0]
X2[:,1]=dfavg.iloc[:,1]

y2=dfavg.iloc[:,2]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import mean_squared_error as MSE


# split data in training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

linreg=LinearRegression()
dt = DecisionTreeRegressor(max_depth=5)
bc=BaggingRegressor(base_estimator=dt,n_estimators=50)

linreg.fit(X2,y2)
dt.fit(X2,y2)
bc.fit(X2,y2)

y_pred_lin=linreg.predict(X_test)
y_pred_tree=dt.predict(X_test)
y_pred_bc=bc.predict(X_test)

print('tree '+str(MSE(y_pred_tree,y_test)**(1/2)))
print('bagging '+str(MSE(y_pred_bc,y_test)**(1/2)))
print('linear '+str(MSE(y_pred_lin,y_test)**(1/2)))

print('errore %'+str(np.mean(np.abs((y_pred_tree-y_test)/y_test)*100)))
```
